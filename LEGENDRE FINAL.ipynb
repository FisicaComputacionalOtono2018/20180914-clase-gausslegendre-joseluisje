```
#ALUMNO: JOSE LUIS JIMENEZ ESPINOZA
#FECHA: 13/09/18
#PROGRAMA QUE RESUELVE EL POLINOMIO DE LEGENDRE DE GRADO ENTERO POSITIVO n Y LO EVALÚA EN UN VALOR REAL x

#VERSION 2 (SIN FUNCIONES ESPECIALES)
import numpy as np
import matplotlib.pyplot as plt
import math
from fractions import Fraction

print("--PROGRAMA QUE RESUELVE EL POLINOMIO DE LEGENDRE DE GRADO ENTERO POSITIVO n Y LO EVALÚA EN UN VALOR REAL x")

n=int(input("Inserte grado del polinomio (n): "))
xv=float(input("Inserte valor a evaluar (x): "))

b=float(2)/1000

dominio=np.arange(-1,1,b)

d=math.factorial(n)
e=2**n
ee=d*e
fracc=s = Fraction(1, ee)

p = np.poly1d([1, 0, -1]) #polinomio inicial
g=p
h=1
while h<n: #n-esima potencia del polinomio
  g= g*p
  h=h+1  
  
k=g
j=0
while j<n: #n-esima derivada del polinomio elevado a la n-esima
  k=np.polyder(k)
  j=j+1 

polfinal=fracc*k #multiplicacion de la fraccion por el polinomio
print("\nEl polinomio es: \n")
print(polfinal)

fxv=polfinal(xv) #polinomio final evaluado en x
print("\nEl polinomio evaluado en x es: ",fxv)

imagen=[] #imagenes para x en el dominio 
for i in dominio:
  fx=polfinal(i)
  imagen.append(fx)
  
plt.plot(dominio,imagen, 'b')

plt.title("\n\nPOLINOMIO DE LEGENDRE PARA n="+str(n)+"\n ")
plt.grid(color='lightgrey', linestyle='-', linewidth=2)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```
