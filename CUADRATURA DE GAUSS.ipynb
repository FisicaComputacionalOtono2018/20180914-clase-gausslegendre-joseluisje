```
#VERSION FINAL
#AUTOR: Jose Luis Jimenez Espinoza
#15/09/18

import numpy as np
import math

print("--PROGRAMA QUE INTEGRA NUMERICAMENTE CON EL METODO DE LA CUADRATURA DE GAUSS LA FUNCION f(x)=x3+2x2 DADOS LOS LIMITES DE INTEGRACION INFERIOR Y SUPERIOR Y EL NUMERO DE NODOS.")

a = float(input("\nInserte el limite de integracion inferior: "))
b = float(input("Inserte el limite de integracion superior: "))
n1 = int(input("Inserte el numero de puntos (esto es igual al grado del polinomio de legendre a utilizar en la cuadratura), debe ser un numero natural: "))

#---//Polinomio de legendre de grado n

def legendre(n):
  
  factorialn = math.factorial(n)
  potn = 2**n
  factor = float(1)/(factorialn*potn)

  #Polinomio inicial
  
  polinomio = np.poly1d([1, 0, -1]) 
  
  #Potencia n-esima del polinomio inicial
  
  potpolinomio = polinomio
  count = 1
  while count<n: 
    potpolinomio = potpolinomio*polinomio
    count = count+1  
  
  #Derivada n-esima del polinomio inicial elevado a la n-esima
  
  derpolinomio = potpolinomio
  count2 = 0
  while count2<n: 
    derpolinomio = np.polyder(derpolinomio)
    count2 = count2+1 
  
  #Polinomio final (Polinomio de legendre de grado n)
  
  pollegendre = factor*derpolinomio 

  return pollegendre

#---//Peso (Wi) de x en funcion de n (grado del polinomio de legendre)

def peso(n,x):

  polinomio=legendre(n)

  #Obtención del peso de x utlizando el polinomio de legendre
  
  derpolfinal = np.polyder(polinomio) 
  derpolfinal2 = derpolfinal**2
  peso = float(2)/((1-(x**2))*(derpolfinal2(x)))
  
  return peso

#---//Nodos (xi) en un vector en funcion de n (grado del polinomio de legendre)

def nodos(n):

  polinomio=legendre(n)
  
  nodos = np.roots(polinomio) #Raíces del polinomio de legendre de grado n
  
  return nodos

#---//Funcion a integrar

def f(x):
  
  funcion = x**3+2*x**2
  
  return funcion

#---//Cuadratura de Gauss

def cuadratura(n,a,b):
  
  q1 = float(b-a)/2
  q2 = float(a+b)/2
  xi = nodos(n)

  pesosf = []
  for i in xi:
    f1 = f(q1*float(i)+q2)
    p1 = f1*peso(n,i)
    pesosf.append(p1)
  sumatoria = sum(pesosf)
  integral = sumatoria*float(q1)
  
  return integral

#---//Resultado final (integral exacta)

integral = cuadratura(n1,a,b)
print("\nEl resultado es: ",integral)


```
